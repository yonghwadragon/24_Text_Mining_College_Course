```python
from konlpy.tag import Okt
t = Okt()
t.nouns(text1) # nouns는 명사만 뽑는다.
```

```python
# data에서 3~5번 리뷰에서 대해서 명사만 각각 뽑으세요.
tmp = data.document.tolist()[2:5]

print(tmp)

print(t.nouns(tmp[0])) # 하나씩 하는 방법
print(t.nouns(tmp[1]))
print(t.nouns(tmp[2]))

for j in tmp:              # 한 번에 하는 방법
  print(t.nouns(j))

[ t.nouns(i) for i in tmp ]  # 한번에 하고 리스트로 묶어주기도 함. 제일 좋은 방법
```
