2. 웹 문서 Parsing   # 파싱시험에 나옴

```python
#Parsing을 위한 라이브러리
!pip install beautifulsoup4
```

```python
import requests
from bs4 import BeautifulSoup     # 아름다운 국 정도는 외우기

url = 'https://comic.naver.com/webtoon/weekday.nhn'
res = requests.get(url)
res.text
```

```python
soup = BeautifulSoup(res.text, "html.parser")
soup
```

위에 바디는 자바스크립트로 동적을 으로 작동해서 셀레니움인가 그거 써야함. 참고임.

```python
soup.head.title
```

3. pandas의 read_html 이용

```python
import pandas as pd
tables = pd.read_html("https://finance.naver.com/item/main.nhn?code=005930", encoding="cp949")  # 인코딩하면 한글도 되도록 하는 것입니다. 표현하는 것을 약속?

#read_html
# match: 문자열, 또는 정규표현식으로 원하는 값이 있는 테이블 선택
# flavor: html 파서 선택, None(선택 시 lxml로 우선 파싱) / ‘bs4’ / ‘html5lib’
# header: 컬럼명으로 사용할 행 지정
# encoding: utf-8, 한글 깨진 경우, cp949
```
